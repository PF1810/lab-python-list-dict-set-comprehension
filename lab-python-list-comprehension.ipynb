````python
#1. 
# Define the functions with comprehensions

def initialize_inventory(products):
    """
    Initialize the inventory dictionary using a dictionary comprehension.
    
    Args:
        products (list): A list of product names.
    
    Returns:
        dict: A dictionary with product names as keys and their quantities as values.
    """
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

def get_customer_orders(products):
    """
    Prompt the user to enter the product names using a set comprehension and return the customer_orders set.
    
    Args:
        products (list): A list of all available product names.
    
    Returns:
        set: A set of product names that the customer wants to order.
    """
    print("Please enter the name of products to order (separated by spaces):")
    customer_orders = {product.lower() for product in input().split() if product.lower() in products}
    return customer_orders

def update_inventory(customer_orders, inventory):
    """
    Update the inventory based on the customer orders using a dictionary comprehension.
    
    Args:
        customer_orders (set): A set of product names that the customer wants to order.
        inventory (dict): A dictionary with product names as keys and their quantities as values.
    
    Returns:
        dict: The updated inventory dictionary.
    """
    updated_inventory = {product: inventory.get(product, 0) - 1 for product in customer_orders}
    return {product: quantity for product, quantity in updated_inventory.items() if quantity >= 0}

def print_updated_inventory(inventory):
    """
    Print the updated inventory.
    
    Args:
        inventory (dict): A dictionary with product names as keys and their quantities as values.
    """
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

products = ["t-shirt", "mug", "hat", "book", "keychain"]

#Initialize inventory with comprehension
inventory = initialize_inventory(products)

#Get customer orders with comprehension
customer_orders = get_customer_orders(products)

#Update inventory based on customer orders with comprehension
inventory = update_inventory(customer_orders, inventory)

#Print updated inventory
print_updated_inventory(inventory)
````

````python
#5
def initialize_inventory(products):
    """
    Initialize the inventory dictionary using a dictionary comprehension.
    
    Args:
        products (list): A list of product names.
    
    Returns:
        dict: A dictionary with product names as keys and their quantities as values.
    """
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

def get_customer_orders(products):
    """
    Prompt the user to enter the product names using a loop and return the customer_orders set.
    
    Args:
        products (list): A list of all available product names.
    
    Returns:
        set: A set of product names that the customer wants to order.
    """
    num_orders = int(input("Enter the number of customer orders: "))
    customer_orders = {input(f"Enter product {i + 1}: ").lower() for i in range(num_orders) if input(f"Enter product {i + 1}: ").lower() in products}
    
    return customer_orders

def update_inventory(customer_orders, inventory):
    """
    Update the inventory based on the customer orders and remove products with zero quantity.
    
    Args:
        customer_orders (set): A set of product names that the customer wants to order.
        inventory (dict): A dictionary with product names as keys and their quantities as values.
    
    Returns:
        dict: The updated inventory dictionary with products having quantity greater than zero.
    """
    updated_inventory = {product: inventory.get(product, 0) - 1 for product in customer_orders}
    # Filter out products with quantity zero or less
    inventory = {product: quantity for product, quantity in updated_inventory.items() if quantity > 0}
    return inventory

def calculate_total_price(customer_orders):
    """
    Calculate the total price of the customer order based on user input for each product price.
    
    Args:
        customer_orders (set): A set of product names that the customer wants to order.
    
    Returns:
        float: The total price of the customer order.
    """
    total_price = sum(float(input(f"Enter the price of {product}: ")) for product in customer_orders)
    return total_price

def print_order_statistics(total_products_ordered, percentage_ordered):
    """
    Print the order statistics.
    
    Args:
        total_products_ordered (int): Total number of products ordered.
        percentage_ordered (float): Percentage of unique products ordered.
    """
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Unique Products Ordered: {percentage_ordered}%")

def print_updated_inventory(inventory):
    """
    Print the updated inventory.
    
    Args:
        inventory (dict): A dictionary with product names as keys and their quantities as values.
    """
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Initialize inventory
inventory = initialize_inventory(products)

# Get customer orders
customer_orders = get_customer_orders(products)

# Update inventory
inventory = update_inventory(customer_orders, inventory)

# Print updated inventory
print_updated_inventory(inventory)

# Calculate total price
total_price = calculate_total_price(customer_orders)
print(f"\nTotal Price: ${total_price:.2f}")

# Calculate order statistics
total_products_ordered = len(customer_orders)
percentage_ordered = (len(customer_orders) / len(products)) * 100

# Print order statistics
print_order_statistics(total_products_ordered, percentage_ordered)
````
